```python
"""
Hierarchical GPU Logistic Regression  (upsample + downsample)
=============================================================
‚Ä¢ Chunk-wise CSV ingestion   ‚Üí memory-safe
‚Ä¢ Stage-1: per-chunk Benign undersample ‚Üí 1:1 (Benign vs Attack)
‚Ä¢ Stage-2: optional per-class balance  ‚Üí 14-class attack
‚Ä¢ NEW  ‚ë† upsample minority ‚ë° downsample majority
"""

# ÊåÇËΩΩ Google Drive
from google.colab import drive
import os
drive.mount('/content/drive')

from pathlib import Path
import cudf, cupy as cp, numpy as np, pandas as pd
from cuml.model_selection import train_test_split
from cuml.preprocessing import StandardScaler
from cuml.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# ========= ÊåáÂÆöÊï∞ÊçÆÈõÜÊñá‰ª∂‰ΩçÁΩÆ =========
DATA_DIR = Path("/content/drive/My Drive/Colab Notebooks/MachineLearningCVE")

CHUNK_ROWS      = 500_000
RNG_SEED        = 42
STAGE2_SAMPLE   = True    # ÊòØÂê¶Âú® Stage-2 ÂÜçÂÅöÂùáË°°

MIN_PER_CLASS   = 1_000   # Á®ÄÊúâÁ±ªËã• <1k ÂÖàÂ§çÂà∂Âà∞ 1k
MAX_PER_CLASS   = 3000   # Â§ßÊà∑Á±ªÊúÄÂ§ö‰øùÁïô 5k
RATIO_CAP       = 20      # ÊàñËÄÖ > (min_cnt √ó 20) ‰πüÁÆóÂ§ßÊà∑
# ===================================


# ---------- helpers ----------
def clean_label(s: str) -> str:
    return "UNKNOWN" if (s is None or pd.isna(s) or str(s).strip() == "") \
        else str(s).replace("‚Äì", "-").replace("‚Äî", "-").strip()

def gpu_gc(tag=""):
    cp.get_default_memory_pool().free_all_blocks()
    if tag:
        print(f"[GPU pool cleared] {tag}")

def upsample_minority(df: pd.DataFrame,
                      min_per_cls: int,
                      seed: int) -> pd.DataFrame:
    vc = df["Label"].value_counts()
    minors = vc[vc < min_per_cls].index.tolist()
    if not minors:
        print("üü¢  No minority classes need up-sampling")
        return df

    rng = np.random.default_rng(seed)
    up_parts = []
    for cls in minors:
        g = df[df["Label"] == cls]
        need = min_per_cls - len(g)
        dup  = g.sample(n=need, replace=True,
                        random_state=int(rng.integers(1e9)))
        up_parts.append(pd.concat([g, dup], ignore_index=True))
        print(f"üî∫  Up-sampled {cls:25s}: {len(g):>5} ‚ûú {min_per_cls} (+{need})")
    keep_df = df[~df["Label"].isin(minors)]
    return pd.concat([keep_df, *up_parts], ignore_index=True)

def auto_downsample(df: pd.DataFrame,
                    max_per_cls: int,
                    ratio_cap: int,
                    seed: int) -> pd.DataFrame:
    vc = df["Label"].value_counts()
    min_cnt = vc.min() or 1

    # Â§ßÊà∑Âà§ÂÆöÔºöÂè™Ë¶ÅÊª°Ë∂≥‰ªª‰∏ÄÊù°‰ª∂Âç≥ÂèØ
    majors = vc[(vc > max_per_cls) | (vc > min_cnt * ratio_cap)].index.tolist()
    if not majors:
        print("üü¢  No majority classes detected ‚Äî skip down-sampling")
        return df

    big_df  = df[df["Label"].isin(majors)]
    rest_df = df[~df["Label"].isin(majors)]
    before  = big_df["Label"].value_counts()

    target_n = min(max_per_cls, int(min_cnt * ratio_cap))
    big_down = (
        big_df.groupby("Label", group_keys=False)
              .apply(lambda g: g.sample(n=target_n, random_state=seed))
    )

    after = big_down["Label"].value_counts()
    print(f"üîª  Auto down-sampling majors (cap={target_n:,} rows each):")
    for cls in majors:
        print(f"    ‚Ä¢ {cls:25s}: {before[cls]:>7,} ‚ûú {after[cls]:>7,} "
              f"(‚àí{before[cls]-after[cls]:,})")
    return pd.concat([big_down, rest_df], ignore_index=True)


# ---------- 0. discover CSV ----------
csv_files = sorted(DATA_DIR.glob("*.csv"))
print(f"üìÇ  Found {len(csv_files)} CSV files, loading ...")

# ---------- 1. load & per-chunk Benign undersample ----------
rng = np.random.RandomState(RNG_SEED)
chunks = []

for csv in csv_files:
    for part in pd.read_csv(csv, chunksize=CHUNK_ROWS):
        part.columns = part.columns.str.strip()
        part = part.drop(columns=[c for c in part.columns
                                  if c.lower() in ["flow id", "source ip",
                                                   "destination ip", "timestamp"]],
                         errors="ignore")

        y_raw = part.pop("Label").apply(clean_label)
        mask_attack = y_raw.ne("BENIGN")
        n_attack = mask_attack.sum()

        mask_benign = y_raw.eq("BENIGN").sample(
            n=min(n_attack, mask_attack.size - n_attack),
            random_state=rng
        )

        keep_mask = mask_attack | mask_benign
        kept = part.loc[keep_mask].reset_index(drop=True)
        kept["Label"] = y_raw[keep_mask].values
        chunks.append(kept)

print(f"‚úÖ  Loaded & balanced chunks: {len(chunks):,}")
cpu_df = pd.concat(chunks, ignore_index=True)
del chunks

# ---------- 2. sampling ----------
print("\nüöÄ  Sampling Phase")

# ÊõøÊç¢‰π±Á†ÅÊ†áÁ≠æ
cpu_df["Label"] = cpu_df["Label"].astype(str)
cpu_df["Label"] = cpu_df["Label"].str.replace('Web Attack ÔøΩ Brute Force', 'Web Attack - Brute Force', regex=False)
cpu_df["Label"] = cpu_df["Label"].str.replace('Web Attack ÔøΩ XSS', 'Web Attack - XSS', regex=False)
cpu_df["Label"] = cpu_df["Label"].str.replace('Web Attack ÔøΩ Sql Injection', 'Web Attack - Sql Injection', regex=False)

cpu_df = upsample_minority(cpu_df, MIN_PER_CLASS, RNG_SEED)
cpu_df = auto_downsample(cpu_df, MAX_PER_CLASS, RATIO_CAP, RNG_SEED)

print("\nüìä  Class distribution AFTER sampling:")
print(cpu_df["Label"].value_counts().sort_values(ascending=False).to_string())
rows, cols = cpu_df.shape
print(f"\nüìê  Data matrix ‚ûú {rows:,} rows √ó {cols} columns "
      f"(features = {cols - 2})\n")   # -2 ÊéíÈô§ Label & LabelIdx

# ---------- 3. label mapping ----------
classes_all = sorted(cpu_df["Label"].unique())
lab2idx = {l: i for i, l in enumerate(classes_all)}
idx2lab = {i: l for l, i in lab2idx.items()}
cpu_df["LabelIdx"] = cpu_df["Label"].map(lab2idx).astype("int32")

# ---------- 4. move to GPU ----------
feat_cols = [c for c in cpu_df.columns if c not in ["Label", "LabelIdx"]]
gdf = cudf.DataFrame(cpu_df[feat_cols].astype("float32"))
gdf = gdf.fillna(0).replace([cp.inf, -cp.inf], 0)
y_codes = cudf.Series(cpu_df["LabelIdx"].values)
del cpu_df

# ---------- 5. train/test split ----------
X_tr, X_te, y_tr, y_te = train_test_split(
    gdf, y_codes, test_size=0.2, stratify=y_codes, random_state=RNG_SEED
)
print(f"‚úÇÔ∏è  Train/Test sizes ‚ûú Train {len(X_tr):,} | Test {len(X_te):,}")

# ---------- 6. scale ----------
scaler = StandardScaler(with_mean=False)
scaler.fit(X_tr)
X_tr_s = scaler.transform(X_tr)
X_te_s = scaler.transform(X_te)
gpu_gc("after scaling")

# ---------- 7. Stage-1 (binary) ----------
is_benign_tr = (y_tr == lab2idx["BENIGN"])
bin_clf = LogisticRegression(max_iter=1200, solver="qn")
bin_clf.fit(X_tr_s, (~is_benign_tr).astype("int32"))
gpu_gc("after Stage-1")

# ---------- 8. Stage-2 (14-class) ----------
mask_att_tr = ~is_benign_tr
X_att, y_att = X_tr[mask_att_tr], y_tr[mask_att_tr]

if STAGE2_SAMPLE:
    counts = y_att.value_counts()
    k_min  = int(max(counts.min(), MIN_PER_CLASS))
    cp.random.seed(RNG_SEED)
    keep_idx = cp.concatenate([
        cp.random.choice(cp.where((y_att == int(cid)).to_cupy())[0],
                         size=min(k_min, int(counts[int(cid)])),
                         replace=False)
        for cid in counts.index.values_host
    ])
    X_att, y_att = X_att.iloc[keep_idx], y_att.iloc[keep_idx]

att_clf = LogisticRegression(max_iter=1200, solver="qn")
att_clf.fit(scaler.transform(X_att), y_att)
gpu_gc("after Stage-2")

# ---------- 9. inference ----------
prob_attack = bin_clf.predict_proba(X_te_s).iloc[:, 1]
flag_attack = (prob_attack >= 0.5).astype("bool")

rows_attack = flag_attack.values_host
att_pred_idx = (att_clf.predict(X_te_s[rows_attack]).to_numpy()
                if rows_attack.any() else np.array([], dtype="int32"))

final_preds, i_att = [], 0
for atk in rows_attack:
    if not atk:
        final_preds.append("BENIGN")
    else:
        final_preds.append(
            idx2lab[int(att_pred_idx[i_att])] if i_att < len(att_pred_idx)
            else "UNKNOWN"
        )
        i_att += 1

y_true = pd.Series([idx2lab[int(i)] for i in y_te.to_numpy()], dtype="object")
y_pred = pd.Series(final_preds, dtype="object")

# ---------- 10. evaluation ----------


# ========= 11. PR-Curve / Per-Class / Confusion =========
print("\nüìà  Generating PR-curve, per-class metrics & confusion matrix ...")

from sklearn.metrics import precision_recall_curve, average_precision_score, classification_report
from pathlib import Path

OUT = Path("./viz_outputs")         # ËæìÂá∫ÁõÆÂΩï
OUT.mkdir(exist_ok=True)

# ---------- 11-1  ÂæÆÂπ≥Âùá PR curve ----------
attack_classes = [c for c in classes_all if c != "BENIGN"]          # 14 attack Á±ª
y_scores = att_clf.predict_proba(X_te_s).to_pandas().values         # (n,14)
y_bin    = (pd.get_dummies(y_true)
              .reindex(columns=attack_classes, fill_value=0)
              .values)

prec, rec, _ = precision_recall_curve(y_bin.ravel(), y_scores.ravel())
ap  = average_precision_score(y_bin, y_scores, average="micro")

plt.figure(figsize=(6,5))
plt.plot(rec, prec, lw=2)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title(f"Micro-avg PR curve (AP={ap:.3f})")
plt.grid(alpha=.3); plt.tight_layout()
plt.savefig(OUT/"pr_curve_micro.png", dpi=150); plt.show()

pd.DataFrame({"recall":rec, "precision":prec})\
  .to_csv(OUT/"pr_curve_micro.csv", index=False)
pd.DataFrame({"recall": rec, "precision": prec})\
  .to_csv(OUT / "pr_curve_micro.csv", index=False)

# ‚¨áÔ∏é Êñ∞Â¢ûÔºöÊâìÂç∞Ââç 10 Ë°åÊàñÂÖ®ÈÉ®Ë°å
print("\nüìÑ  Micro-PR  dataÔºö")
print(pd.read_csv(OUT / "pr_curve_micro.csv"))
# Ëã•ÊÉ≥ÂÖ®ÈÉ®ÊâìÂç∞Ôºöprint(pd.read_csv(...).to_string(index=False))

# ---------- 11-2  ÊØèÁ±ª Precision / Recall ----------
rep = classification_report(y_true, y_pred,
                            target_names=classes_all, output_dict=True)
cls_df = (pd.DataFrame(rep).T
            .loc[classes_all, ["precision","recall","f1-score","support"]]
            .round(4))
cls_df.to_csv(OUT/"per_class_metrics.csv")

x = np.arange(len(classes_all)); w=.35
plt.figure(figsize=(12,5))
plt.bar(x-w/2, cls_df["precision"], w, label="Precision")
plt.bar(x+w/2, cls_df["recall"],    w, label="Recall")
plt.xticks(x, classes_all, rotation=90, fontsize=8)
plt.ylim(0,1); plt.ylabel("Score")
plt.title("Per-class Precision & Recall"); plt.legend()
plt.tight_layout(); plt.savefig(OUT/"per_class_prec_recall.png", dpi=150); plt.show()

print("\nüóíÔ∏è  Per-class metrics (topÂá†Ë°åÈ¢ÑËßà)Ôºö")
display(cls_df)

# ---------- 11-3  Confusion matrix ----------
print("\n=== Hierarchical metrics ===")
print(classification_report(y_true, y_pred, digits=3))
print(f"üéØ  Test Accuracy: {accuracy_score(y_true, y_pred):.4f}")

order = ["BENIGN"] + [c for c in classes_all if c != "BENIGN"] + ["UNKNOWN"]
cm = confusion_matrix(y_true, y_pred, labels=order)
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(12, 9))
plt.imshow(cm_norm, cmap="Blues")
plt.title("Hierarchical 15-class Confusion Matrix (%)")
plt.colorbar(label="Percent")
plt.xticks(np.arange(len(order)), order, rotation=90, fontsize=7)
plt.yticks(np.arange(len(order)), order, fontsize=7)
thr = cm_norm.max() / 2
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        v = cm_norm[i, j]
        if v > 0.001:
            plt.text(j, i, f"{v*100:.1f}%", ha="center", va="center",
                     color="white" if v > thr else "black", fontsize=5)
plt.tight_layout()
plt.show()


# ------- (Â∑≤Áªè‰øùÂ≠ò micro-PR) -------

# ---------- 11-1-bis  ÊØè‰∏™ attack Á±ªÁã¨Á´ã PR ----------
print("\nüìÑ  Per-class PR-curve Êï∞ÊçÆ & AP ÂÄºÔºö")
for i, cls in enumerate(attack_classes):          # attack_classes = classes_all[1:]
    p, r, _ = precision_recall_curve(y_bin[:, i], y_scores[:, i])
    ap_cls  = average_precision_score(y_bin[:, i], y_scores[:, i])
    cls_df  = pd.DataFrame({"recall": r, "precision": p})
    csv_path = OUT / f"pr_curve_{cls.replace(' ', '_')}.csv"
    cls_df.to_csv(csv_path, index=False)
    print(f"  ‚Ä¢ {cls:<25s}  AP={ap_cls:.3f}   ‚ûú  {csv_path.name}")
```
